# Import libraries and datsets

```python
# import libraries
import os
import pandas as pd
import matplotlib.pyplot as plt
import statistics as stat
import seaborn as sns
```

```python
# Function to create horizontal bar plot for top 10 GO terms
def plot_top10_go_terms(df, title):
    # Sort the dataframe by adjusted p-value
    df_sorted = df.sort_values(by='Adjusted P-value').head(10)

     # Function to insert a newline if the label length is greater than 20 characters
    def wrap_labels(label, max_length=70):
        return '\n'.join([label[i:i+max_length] for i in range(0, len(label), max_length)])

    # Apply the function to wrap long labels
    df_sorted['Term'] = df_sorted['Term'].apply(lambda x: wrap_labels(x, 70))

    # Create the bar plot
    plt.figure(figsize=(10, 6))
    barplot = sns.barplot(
        x='Adjusted P-value',
        y='Term',
        data=df_sorted,
        palette='viridis',
        edgecolor='w'
    )

    # Add title and labels
    plt.title(title)
    plt.xlabel('Adjusted P-value')
    plt.ylabel('GO Term')

    # Adjust layout to accommodate labels and titles
    plt.subplots_adjust(left=0.3, right=0.95, top=0.9, bottom=0.1)

    # Display the plot
    #plt.tight_layout()
    plt.show()
```

```python
# Load the GO enrichment table
file_path = 'Go insilico favorable.csv'
df = pd.read_csv(file_path)

# Plot the top 10 GO terms
plot_top10_go_terms(df, 'Top 10 GO Terms - In Silico Favorable')
```

```python
# Load the GO enrichment table
file_path = 'Go insilico unfavorable.csv'
df = pd.read_csv(file_path)

# Plot the top 10 GO terms
plot_top10_go_terms(df, 'Top 10 GO Terms - In Silico Unfavorable')
```
