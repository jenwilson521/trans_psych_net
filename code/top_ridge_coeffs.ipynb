```python
import pandas as pd
import os
from collections import defaultdict
from numpy import nan
```

```python
path_in_silico = 'insilico_phagocyt_ridgeregcoeffs.xlsx'
in_silico_ridgecoeffs = pd.read_excel(path_in_silico)
in_silico_ridgecoeffs
```

```python
import matplotlib.pyplot as plt

# Sort the data by 'Regression Coefficients'
sorted_df = in_silico_ridgecoeffs.sort_values(by='Regression Coefficients')

# Select the bottom 20 and top 20
bottom_20 = sorted_df.head(20)
top_20 = sorted_df.tail(20)

# Combine the two DataFrames
combined_df = pd.concat([bottom_20, top_20])

# Plot the bar graph
plt.figure(figsize=(14, 7))
plt.bar(combined_df['Gene Names'], combined_df['Regression Coefficients'], color=['blue' if x < 0 else 'orange' for x in combined_df['Regression Coefficients']])
plt.xticks(rotation=90)
plt.xlabel('Gene Names')
plt.ylabel('Regression Coefficients')
plt.title('Top 20 and Bottom 20 In Silico Ridge Regression Coefficients')
plt.show()
```
