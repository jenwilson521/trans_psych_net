```python
import matplotlib.pyplot as plt

# Data for the bar graph
categories = ['Unique Drug-Phenotype Pairs', 'Unique Drugs', 'Unique Phenotypes']
counts = [2245, 757, 4]

# Create the bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, counts, color='purple')

# Add title and labels
plt.title('Drug and Phenotype Counts for PathFX Analysis', fontsize=14)
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Counts', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Display the counts on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha='center', va='bottom', fontsize=14)

# Save figure as png
plt.savefig("remake.png", dpi=300)

# Show the plot
plt.tight_layout()
plt.show()
```
